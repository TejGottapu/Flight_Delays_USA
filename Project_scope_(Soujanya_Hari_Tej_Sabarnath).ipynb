```python
import pandas as pd
data = pd.read_csv("Flight Delays.csv")
data.head()
```

```python
%matplotlib inline
from matplotlib import pyplot as plt
```

```python
# Below data is a snippet of only the first 100,000 values of overall records captured in the chosen Dataset.

data = data.head(100000)
```

```python
## set the figure dimensions
hist_fig = plt.figure(figsize = (6, 6))
    
## make a histogram and set the number of bins to 15
_  = plt.hist(data["DAY_OF_WEEK"], bins = 15)

## Set the tick and label fontsize
plt.tick_params(labelsize = 15)

## Set the title
_ = plt.title("Number of Flight delays Vs Day_of_week", fontsize = 15)

## Set the y-label
_ = plt.ylabel("Number of Flight delays", fontsize = 15)

## Set the y-label
_ = plt.xlabel("DAY_OF_WEEK", fontsize = 15)

#_ = plt.legend(fontsize = 15)
```

```python
## set the figure dimensions
import scipy.stats
hexb_fig = plt.figure(figsize = (6, 6))

## To put a (Pearson) correlation into a figure
## we must process the relationship between the two
## variables and save the output as a string:

cor = scipy.stats.pearsonr(
    data["ARRIVAL_DELAY"],data["DAY_OF_WEEK"]
)[0]

## for clarity, it's probably best to round
## to a few decimal places, first
## but we should also label it as "r"
corstr = "r = " + str(round(cor, 2))
    
## make a hexbin plot
_  = plt.hexbin(
    data["ARRIVAL_DELAY"],data["DAY_OF_WEEK"], color = "black",
    gridsize = 30, cmap = "binary", label = corstr
)

## Set the tick and label fontsize
plt.tick_params(labelsize = 15)

## Set the title
_ = plt.title("ARRIVAL_DELAY Vs. DAY_OF_WEEK", fontsize = 15)

## Set the y-label
_ = plt.ylabel("ARRIVAL_DELAY", fontsize = 15)

## Set the y-label
_ = plt.xlabel("DAY_OF_WEEK", fontsize = 15)

## make a legend
_ = plt.legend(fontsize = 15)
```

```python
scat_fig = plt.figure(figsize = (6, 8))
    
## make a scatter plot
_  = plt.scatter(
    data["AIRLINE"], data["DAY"],
    color = "blue"
)

## Set the tick and label fontsize
plt.tick_params(labelsize = 15)

## Set the title
_ = plt.title("DAY Vs. AIRLINE", fontsize = 15)

## Set the y-label
_ = plt.ylabel("DAY", fontsize = 15)

## Set the y-label
_ = plt.xlabel("AIRLINE", fontsize = 15)
```
